```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-10, 10, 101)
```

```python
def ReLU(x):
    data = [max(0,value) for value in x]
    y = np.array(data, dtype=float)
    
    # Graph
    plt.plot(x, y)
    plt.title('ReLU Activation Function')
    plt.grid()
    plt.show()
```

```python
def sigmoid(x):
    data = [(1/(1+np.exp(-value))) for value in x]
    y = np.array(data, dtype=float)

     # Graph
    plt.plot(x, y)
    plt.title('Sigmoid Activation Function')
    plt.grid()
    plt.show()
```

```python
def tanh(x):
    data = [( (np.exp(value) - np.exp(-value)) / (np.exp(value) + np.exp(-value)) ) for value in x]
    y = np.array(data, dtype=float)

     # Graph
    plt.plot(x, y)
    plt.title('Sigmoid Activation Function')
    plt.grid()
    plt.show()
```

```python
tanh(x)
```

```python
sigmoid(x)
```

```python
ReLU(x)
```
